## 1개의 열만 고정하고 나머지 열을 행으로 바꾸기

```python
import pandas as pd
pew = pd.read_csv('../data/pew.csv')
print(pew)
# religion 열은 그대로 두고 나머지 열을 하나의 열에 다 녹여서 표현한다.
# 따라서 나머지 열을 행으로 바꿔진다.
# id_vars는 고정할 열의 이름
# var_name은 원래 열의 이름이 들어가는 열의 이름 설정 값
# value_name은 원래 열에 있던 값이 들어가는 열의 이름 설정 값
pew_long = pd.melt(pew, id_vars='religion', var_name='income', value_name='count')
print(pew_long)
```

## 2개 이상의 열을 고정하고 나머지 열을 행으로 바꾸기

```python
billboard = pd.read_csv('../data/billboard.csv')
print(billboard)

# id_vars에 리스트로 열의 이름들을 넣으면 해당 열들은 고정 된다.
billboard_long = pd.melt(billboard, id_vars=['year', 'artist', 'track', 'time', 'date.entered'], var_name='week', value_name='rating')
print(billboard_long)
```

## ebola 데이터 집합 살펴보기

```python
ebola = pd.read_csv('../data/country_timeseries.csv')
print(ebola.columns)
print(ebola)
```

```python
ebola_long = pd.melt(ebola, id_vars=['Date', 'Day'])
ebola_long
```

```python
print(type(ebola_long.variable))
ebola_long.variable # ebola_long['variable']
```

# 열 이름 나누고 데이터 프레임에 추가하기

```python
# pandas의 str 메소드를 이용하여 '_' 단위로 두개의 열로 나눈다.
# 이때 분리된 형태는 시리즈로 하나의 열을 갖고 있다.
variable_split = ebola_long.variable.str.split('_') 
print(type(variable_split))
print(variable_split)

# 스플릿하여 0 열과 1 열의 결과 값을 직접 데이터 프레임에 추가해준다.
status_values = variable_split.str.get(0) 
country_values = variable_split.str.get(1)
ebola_long['status'] = status_values 
ebola_long['country'] = country_values
ebola_long
```

# concat 메서드를 응용하여 데이터프레임에 열 추가하기

```python
# pandas의 str 메소드를 이용하여 '_' 단위로 두개의 열로 나눈다.
# expand를 사용하여 시리즈 형태로 분리하는 것이 아닌 데이터프레임으로 분리한다.
# 분리된 데이터 프레임은 두개의 열을 갖고있다.
variable_split = ebola_long.variable.str.split('_', expand=True) 
print(variable_split)

# 나눈 열의 이름을 각각 설정해준다.
variable_split.columns = ['status', 'country'] 
print(variable_split)
# 원래 있던 데이터 프레임에 판다스 concat 메소드를 이용하여 열 단위로
# 데이터 프레임에 추가해준다.
ebola_parsed = pd.concat([ebola_long, variable_split], axis=1)
print(ebola_parsed)
```

# 기상 데이터의 여러 열을 하나로 정리하기

```python
weather = pd.read_csv('../data/weather.csv') 
print(weather.iloc[:5, :11])
# 기상 데이터를 4개의 열로 고정하고 나머지 열을 하나로 녹여낸다.
weather_melt = pd.melt(weather, id_vars=['id', 'year', 'month', 'element'], var_name='day', value_name='temp') 
weather_melt
```

```python
# 피벗 테이블 메소드을 통해 인덱스를 계층적으로 묶어서 표현할 수 있다.
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'], 
    columns='element')
weather_tidy
```

```python
# 피벗 테이블의 인덱스 계층 레벨을 리셋 할 수 있다.
weather_tidy_flat = weather_tidy.reset_index() 
weather_tidy_flat
```
